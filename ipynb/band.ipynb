## Using hexagonal unit cell's k-path

```python
from abipy.abilab import abiopen
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.sans-serif'] = "arial"
fig,ax=plt.subplots(figsize=(5,4))
with abiopen("AlScN.nscfo_DS2_GSR.nc") as ncfile:
    band=ncfile.ebands
band.plot(with_gaps=True, title="Band structure", ylims=(-2,4), 
          ax=ax)
plt.show()
```
