```python
import abipy.data as abidata
import abipy.abilab as abilab
from abipy.abilab import AbinitInput
from pymatgen.io.abinit import inputs
from pymatgen.core import Structure

structure=Structure.from_file('./AlN/algan.cif')
# structure[1].species='Ga'
abilab.enable_notebook()
abidata.pseudo_dir = './JTH-PBE-atomicdata-1.1/ATOMICDATA/'
pseudos=[f'{i}.GGA_PBE-JTH.xml' for i in structure.symbol_set]
inp = AbinitInput(structure=structure,
                  pseudos=pseudos, pseudo_dir=abidata.pseudo_dir, )

#Additional parameters
# inp.set_autokmesh(nksmall=1)
inp.set_vars(ecut=150, pawecutdg=150, toldfe=1e-6)
inp.set_kmesh([1,12,12],[0.5,0.5,0.5], 1)
#Write an input file
# inp.write('../au_mos2/abinit/st0.abi')

#View input file as below:
print(inp)
```
